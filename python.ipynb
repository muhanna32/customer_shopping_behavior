```python

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df = df.drop(columns=['promo_code_used'], errors='ignore')
```

```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql

# database configuration
username = "root"
password = "root"  # MySQL root password
host = "127.0.0.1"
port = "3306"
database = "s2"

print("--- Connecting to MySQL Database ---")
try:
    # 1. Connect without database to ensure database exists
    conn = pymysql.connect(host=host, user=username, password=password, port=int(port))
    with conn.cursor() as cursor:
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
    conn.close()
    print(f"Database '{database}' is ready.")
    
    # 2. Connect using SQLAlchemy
    engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
    
    # Convert categorical columns to string for database compatibility
    df_db = df.copy()
    for col in df_db.select_dtypes(include=['category']).columns:
        df_db[col] = df_db[col].astype(str)
        
    table_name = "customer_shopping_behavior"
    print(f"Saving dataframe to table '{table_name}'...")
    df_db.to_sql(table_name, engine, if_exists="replace", index=False)
    print("Data saved successfully!")
    
    # 3. Retrieve and display sample data
    query = f"SELECT * FROM {table_name};"
    df_mysql = pd.read_sql(query, con=engine)
    print("\n--- Data successfully retrieved from MySQL ---")
    print(df_mysql.head())
except Exception as e:
    print(f"\n[ERROR] Connection or SQL operation failed: {e}")
    print("Please make sure MySQL is running and database server is accessible.")
```
